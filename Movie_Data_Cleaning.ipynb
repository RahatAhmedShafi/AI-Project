```python
df['Genres'] = df['Genres'].apply(
    lambda x: [genre.strip().lower() for genre in x.split(',')] if isinstance(x, str) else [genre.strip().lower() for genre in x]
)

df.head(5)
```

```python
df['combined_features'] = df['Movie'] + ' ' + \
                          df['Writer'] + ' ' + \
                          df['Director'] + ' ' + \
                          df['Genres'].apply(lambda x: ' '.join(x)) + ' ' + \
                          df['Certification/Rating']

df.head(5)
```

```python

# df.drop(columns=['IMDb ID'], inplace=True)
# df.head(5)
```

```python

# Fill NaN with empty strings
df['combined_features'] = df['combined_features'].fillna('')

# Now TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

# These seem like great related recommendations based on similar genres, directors, writers, or ratings.

```python

# Indexing by movie title
indices = pd.Series(df.index, index=df['Movie']).drop_duplicates()

# def get_recommendations(title, cosine_sim=cosine_sim):
#     title = title.lower()
#     idx = indices[title]

#     sim_scores = list(enumerate(cosine_sim[idx]))
#     sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:6]

#     movie_indices = [i[0] for i in sim_scores]
#     return df['Movie'].iloc[movie_indices]

def get_recommendations(title, cosine_sim=cosine_sim):
    # Convert to lowercase to handle case sensitivity
    title = title.lower()

    # Create a mapping from lowercase titles to their actual index
    title_indices = pd.Series(df.index, index=df['Movie'].str.lower())

    # Check if the title exists
    if title not in title_indices:
        return f"❌ Movie '{title}' not found in dataset."

    # Get the index of the movie
    idx = title_indices[title]

    # Get similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort and get top 5 similar movies (excluding itself)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:6]

    # Get the indices of recommended movies
    movie_indices = [i[0] for i in sim_scores]

    # Return movie titles
    return df['Movie'].iloc[movie_indices]
```

```python
def get_movie_info_and_recommendations(title, cosine_sim=cosine_sim):
    # Convert input to lowercase
    title = title.lower()

    # Create a case-insensitive index for movie lookup
    title_indices = pd.Series(df.index, index=df['Movie'].str.lower())

    # Check if movie is in dataset
    if title not in title_indices:
        print(f"❌ Movie '{title}' not found in dataset.")
        return

    # Get the movie index
    idx = title_indices[title]

    # 1. Show input movie information
    print("🎬 Input Movie Information:\n")
    info = df.loc[idx][['Movie', 'Year', 'Genres', 'Certification/Rating', 'Writer', 'Director']]
    for label, value in info.items():
        print(f"{label}: {value}")

    # 2. Get similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:6]  # skip itself

    # 3. Get the recommended movie titles
    movie_indices = [i[0] for i in sim_scores]
    recommended_titles = df['Movie'].iloc[movie_indices]

    # 4. Print nicely formatted recommendations
    print("\n🔍 Top 5 Recommended Movies:\n")
    for i, title in enumerate(recommended_titles, 1):
        print(f"{i}. {title}")
```

```python
def get_movie_info_and_recommendations(user_input, df=df, cosine_sim=cosine_sim):
    user_input = user_input.lower()
    title_indices = pd.Series(df.index, index=df['Movie'].str.lower())

    if user_input in title_indices:
        # 🎬 Case 1: Exact movie title found
        idx = title_indices[user_input]

        print("🎬 Input Movie Information:\n")
        info = df.loc[idx][['Movie', 'Year', 'Genres', 'Certification/Rating', 'Writer', 'Director']]
        for label, value in info.items():
            print(f"{label}: {value}")

        # Recommendation based on content similarity
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:6]
        movie_indices = [i[0] for i in sim_scores]
        recommended_titles = df['Movie'].iloc[movie_indices]

        print("\n🔍 Top 5 Recommended Movies:\n")
        for i, title in enumerate(recommended_titles, 1):
            print(f"{i}. {title}")

    else:
        # 🧩 Case 2: Genre or keyword match
        matched_df = df[df['combined_features'].str.contains(user_input, case=False, na=False)]

        if matched_df.empty:
            print(f"❌ No match found for '{user_input}'. Try another title or keyword.")
            return

        print(f"\n🔎 Found {len(matched_df)} movies with '{user_input}':")
        for title in matched_df['Movie']:
            print(f"- {title}")
```
